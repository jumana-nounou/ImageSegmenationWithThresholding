```python
from google.colab.patches import cv2_imshow
import cv2 as cv2
import urllib.request
import numpy as np
import pandas as pd
```

```python
#loading the image
img=cv2.imread("coins.jpg")
```

```python
ROI=img[0:700,0:750]
```

```python
GB = cv2.GaussianBlur(ROI, (5,5), 0)
GB
```

```python
gray=cv2.cvtColor(GB,cv2.COLOR_BGR2GRAY)
```

```python
adaptiveThresh=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
```

```python
cv2_imshow(img)
cv2_imshow(adaptiveThresh)
```
